
Abre el archivo 'sparse_ sample_ monthrg.dat.txt' como lectura

```python
t=np.loadtxt(open('sparse_sample_monthrg.dat.txt','r'))
```

Separa las columnas de año y mes

```python
year1=t[:,0]
month1=t[:,1]
X_norm1=t[:,3]
```

Cuenta el numero de datos que tenga la lista de datos ingresados

```python
numDatosOri1=shape(year1)[0]-1
numDatosOri1
```

Encuentra el ultimo dato que tenga el valor -99 y comienza la cuenta desde alli

```python
posicion=0
for i in (range(numDatosOri1)):
    dato=X_norm1[i]
    if(dato==-99):
        posicion=i
        
posicion   
```

#### Se toman los datos utiles y se alteran las variables

```python
year=empty(0)
month=empty(0)
X_norm=empty(0)
for a in range(posicion+1,numDatosOri1):
    year=np.append(year,year1[a])
    month=np.append(month,month1[a])
    X_norm=np.append(X_norm,X_norm1[a])
numDatosOri=shape(year)[0]-1
```

Halla el numero de datos que contendra el nuevo arreglo entre el primer dato y el ultimo

```python
PosInic=month[0]+year[0]*12
PosFin=month[numDatosOri]+year[numDatosOri]*12
RangoTot=(int)(PosFin-PosInic)
RangoTot
```

Funcion $TransPosFecha(Year,Month)$ ,  que convierte la posicion de la fecha de toma de datos en posicion en el total de datos. Recibe como entrada fecha de la toma de datos

```python
def TransPosFecha(yr,mo):
    var1=yr*12+mo
    var2=var1-PosInic
    return var2

#En caso de necesitarlo... la inversa de la funcion
def iTransPosFecha(num):
    var1=num%12
    var2=(int)(num/12)
    var3=year[0]+var2
    var4=month[0]+var1
    return var3,var4
```

```python
Originaldata=empty(0)
for i in range(numDatosOri+1):
    value=TransPosFecha(year[i],month[i])
    Originaldata=np.append(Originaldata,value)
Originaldata[numDatosOri]
```

```python
#Prueba de lista de datos
Data=empty(0)
for a in range(numDatosOri+1):
    Data=np.append(Data,X_norm[a])
shape(Originaldata),shape(Data)
```

```python
InterList1=empty(0)
for a in range(RangoTot+1):
    guarda=0.0
    
    for b in range(numDatosOri+1):
        var1=TransPosFecha(year[b],month[b])
        if a == var1:
            InterList1=np.append(InterList1,X_norm[b])
    try:
        existencia=InterList1[a]*1
    except IndexError:
        InterList1=np.append(InterList1,InterList1[a-1])
#CheckPoint
```

```python
plot(InterList1)
```

###### 2: Tomar la pendiente entre valores y conseguir los valores intermedios (interpolacion lineal). Nombre: $InterList2$

```python
#metodo B
f1=interpolate.interp1d(Originaldata,Data,'linear')

InterList2=empty(0)
for a in range(RangoTot+1):
    InterList2=np.append(InterList2,f1(a))
```
