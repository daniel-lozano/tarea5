```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
from scipy import interpolate
from scipy.fftpack import fft, fftfreq
```


Abre el archivo 'sparse_ sample_ monthrg.dat.txt' como lectura 
para separar cada columna y trabajar con los datos

```python
t=np.loadtxt(open('sparse_sample_monthrg.dat.txt','r'))
```

Separa las columnas de año y mes

```python
year=t[:,0]
month=t[:,1]
X_norm=t[:,3]
```

Cuenta el numero de datos que tenga la lista de datos ingresados

```python
numDatosOri=shape(year)[0]-1
numDatosOri
```

Halla el numero de datos que contendra el nuevo arreglo entre el primer dato y el ultimo

```python
PosInic=month[0]+year[0]*12
PosFin=month[numDatosOri]+year[numDatosOri]*12
RangoTot=(int)(PosFin-PosInic)
RangoTot
```

Funcion $TransPosFecha(Year,Month)$ ,  que convierte la posicion de la fecha de toma de datos en posicion en el total de datos. Recibe como entrada fecha de la toma de datos

```python
def TransPosFecha(yr,mo):
    var1=yr*12+mo
    var2=var1-PosInic
    return var2

#En caso de necesitarlo... la inversa de la funcion
def iTransPosFecha(num):
    var1=num%12
    var2=(int)(num/12)
    var3=year[0]+var2
    var4=month[0]+var1
    return var3,var4
```

```python
Originaldata=empty(0)
for i in range(numDatosOri+1):
    value=TransPosFecha(year[i],month[i])
    Originaldata=np.append(Originaldata,value)
Originaldata[numDatosOri]
```

```python
#Prueba de lista de datos
Data=empty(0)
for a in range(numDatosOri+1):
    Data=np.append(Data,X_norm[a])
shape(Originaldata),shape(Data)
```

####  Existen varios casos para completar los espacios de fechas:  

###### 1: Tomar el mismo valor hasta una nueva fecha (Interpolacion Constante). Nombre: $InterList1$

```python
InterList1=empty(0)
for a in range(RangoTot+1):
    guarda=0.0
    
    for b in range(numDatosOri+1):
        var1=TransPosFecha(year[b],month[b])
        if a == var1:
            InterList1=np.append(InterList1,X_norm[b])
    try:
        existencia=InterList1[a]*1
    except IndexError:
        InterList1=np.append(InterList1,InterList1[a-1])
#CheckPoint
```

```python
plot(InterList1)
```

###### 2: Tomar la pendiente entre valores y conseguir los valores intermedios (interpolacion lineal). Nombre: $InterList2$

```python
#metodo B
f1=interpolate.interp1d(Originaldata,Data,'linear')

InterList2=empty(0)
for a in range(RangoTot+1):
    InterList2=np.append(InterList2,f1(a))
```

```python
plot(InterList2,'-',InterList1,'.')
xlim(4620,4625)
```

###### 3: Tomar interpolacion cuadratica. Nombre: $InterList3$

Se define la funcion $f$ de interpolacion cuadratica

```python
f=interpolate.interp1d(Originaldata,Data,'cubic')
```

```python
InterList3=empty(0)
for a in range(RangoTot+1):
    InterList3=np.append(InterList3,f(a))
#shape(List),shape(InterList2)[0],shape(InterList1)[0],RangoTot
```

```python
plot(InterList3,'b-',InterList2,'ko',InterList1,'r.')
ylim([-100,250])
xlim([2000,2050])
```

Se encuentra la transformada de fourier de cada arreglo y se hallan los espectros de potencias.

```python
dt=1
freq = fftfreq(RangoTot+1, dt)
freq_shift=np.fft.fftshift(freq)
```

###### Para lista 1

```python
#numero de puntos en el intervalo: RangoTot
#nombre de listas: InterList1 , InterList2 , InterList3
fft_l1 = fft(InterList1)
fft_l1 /= RangoTot

fft_l1_shift= np.fft.fftshift(fft_l1)
```

###### Para Lista 2

```python
fft_l2 = fft(InterList2)
fft_l2 /= RangoTot

fft_l2_shift= np.fft.fftshift(fft_l2)
```

###### Para Lista 3

```python
fft_l3 = fft(InterList3)
fft_l3 /= RangoTot

fft_l3_shift= np.fft.fftshift(fft_l3)
```

##### Espectros de potencias

```python
pot_l1=empty(0)
pot_l2=empty(0)
pot_l3=empty(0)

for a in range(RangoTot+1):
    pot_l1=np.append(pot_l1,np.abs(fft_l1[a])**2)
    pot_l2=np.append(pot_l2,np.abs(fft_l2[a])**2)
    pot_l3=np.append(pot_l3,np.abs(fft_l3[a])**2)
    
pot_l1/=RangoTot
pot_l2/=RangoTot
pot_l3/=RangoTot
```
