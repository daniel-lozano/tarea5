```python
"""
matrices utilizadas para el calculo de fft, valores absolutos, etc 
"""
FFT=np.zeros([400,24],dtype=complex)#FFT contendra las transformadas de fourier de todos los datos
ABS=np.zeros([400,24],dtype=float) #ABS tendra las normas de las posiciones de FFT
ABS_2=np.zeros([400,24],dtype=float) #ABS_2 tendra las posiciones de ABS al cuadrado
FREQ=np.zeros([400,24],dtype=float) #frecuncia contendra las frecuencias de fourier de todos los datos
FREQ_1=np.zeros([199,24],dtype=float) #frecuencias positivas
ABS_21=np.zeros([199,24],dtype=float) #val absolutos correspondientes a FREQ_1
 
dt=1
shape(FREQ)
shape(archivo)[0]
```

```python
"""
    se calcula la transformada de fourirer de todas las senales y se guarda las frecuencias en FREQ1
"""
a=0
for i in range(datos):
    fft_y=fft(archivo[:,i])/shape(archivo)[0]
    FFT[:,i]=fft_y
    fft_z=fftfreq(shape(archivo)[0],dt)
    FREQ[:,i]=fft_z
```

```python
"""
    se crea una matriz con los valores absolutos de cada punto de las transformadas y la matriz ABS_2
"""
for i in range(shape(FFT)[0]):
    for j in range(shape(FFT)[1]):
        ABS[i,j]=np.abs(FFT[i,j])
        
for i in range(shape(ABS)[0]):
    for j in range(shape(ABS)[1]):
        ABS_2[i,j]=np.abs(ABS[i,j])**2
    
med=shape(archivo)[0]
```

```python
"""
se crean las matrices donde se encontraran los valores utiles de frecuencia (mayores que 0) 
y los valores de normas correspondientes
"""
for i in range(datos):
    for j in range(1,ne):
        FREQ_1[j-1,i]=log(FREQ[j,i])
    
for i in range(datos):
    for j in range(1,ne):
        ABS_21[j-1,i]=log(ABS_2[j,i])
```
